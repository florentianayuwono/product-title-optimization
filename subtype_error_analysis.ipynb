```python
import pandas as pd
import os
import numpy as np
```

```python
error_type_df = pd.read_csv(os.path.join('../data', 'subtype_error.csv'))
error_type_df
```

```python
import matplotlib.pyplot as plt

relevance_count = error_type_df['Relevance (1-5)'].value_counts().sort_index()
relevance_count.plot(kind='bar')
plt.xlabel('Relevance rating')
plt.ylabel('Count')
plt.title('Distribution of Relevance')
```

```python
engagement_count = error_type_df['Engagement (1-5)'].value_counts().sort_index()
engagement_count.plot(kind='bar')
plt.xlabel('Engagement rating')
plt.ylabel('Count')
plt.title('Distribution of Engagement')
```

```python
fluency_count = error_type_df['Fluency (1-5)'].value_counts().sort_index()
fluency_count.plot(kind='bar')
plt.xlabel('Fluency rating')
plt.ylabel('Count')
plt.title('Distribution of Fluency')
```

```python
import textwrap

relevance_subtype = ['Wrong keywords', 'Missing information', 'Unrelated information', 'Untranslated and not understandable keywords']
engagement_subtype = ['Lack of product appeal/characteristics', 'Repetition of same keywords', 'Inappropriate choice of words', 'Visual/Typography issue']
fluency_subtype = ['Readability issue', 'Containing foreign phrases', 'Word order issue', 'Stylistic issue']

group_counts = error_type_df[fluency_subtype].apply(pd.Series.value_counts)
group_counts = group_counts.T  # Transpose the DataFrame

# Wrap long labels
wrapped_labels = [textwrap.fill(label, width=20) for label in group_counts.index]

ax = group_counts.plot(kind='barh', stacked=True, legend=False)  # Use 'barh' for horizontal bar chart

# Set the y-axis labels with the wrapped text
ax.set_yticklabels(wrapped_labels)

plt.ylabel('Error Sub-Type')  # Set the y-axis label
plt.xlabel('Count')  # Set the x-axis label
plt.title('Fluency')

plt.tight_layout()  # Adjust the plot layout to prevent label overlapping
plt.legend(title='Values', bbox_to_anchor=(1, 1))
plt.show()
```
