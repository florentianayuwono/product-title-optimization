```python
# Import packages
import pandas as pd
import os
import numpy as np
```

```python
# Load data
local_df = pd.read_csv(os.path.join('../data', 'accessories_local.csv'))
local_df = local_df.drop(['shop_id', 'item_id', 'source', 'level2_global_be_category'], axis=1)
# local_df = local_df[local_df['total_sales'] > 0]
# local_df = local_df[local_df['price'] < 100000]

cb_df = pd.read_csv(os.path.join('../data', 'accessories_cb.csv'))
cb_df = cb_df.drop(['shop_id', 'item_id', 'source', 'level2_global_be_category'], axis=1)
# cb_df = cb_df[cb_df['total_sales'] > 0]
# cb_df = cb_df[cb_df['price'] < 100000]
```

```python
cb_df
```

```python
local_total_sales = local_df['total_sales'].values
cb_total_sales = cb_df['total_sales'].values

local_avg_price = local_df['price'].mean()
cb_avg_price = cb_df['price'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 4))

# Create boxplots to identify outliers
sns.boxplot(data=[local_total_sales, cb_total_sales])
plt.xticks([0, 1], ['Local', 'Cross-Border'])
plt.xlabel('Product Type')
plt.ylabel('Total Sales')
plt.title('Identification of Outliers')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.kdeplot(local_total_sales, fill=True, label='Local')
sns.kdeplot(cb_total_sales, fill=True, label='Cross-Border')

# Set plot labels
plt.xlabel('Total Sales')
plt.ylabel('Density')
plt.title('Distribution of Total Sales')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=local_df, x='price', y='total_sales')

# Set plot labels
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Correlation between Price and Total Sales')

# Display the plot
plt.show()

correlation_coef = local_df['price'].corr(local_df['total_sales'])
print("Pearson correlation coeff:", correlation_coef)
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=cb_df, x='price', y='total_sales')

# Set plot labels
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Correlation between Price and Total Sales')

# Display the plot
plt.show()

correlation_coef = cb_df['price'].corr(cb_df['total_sales'])
print("Pearson correlation coeff:", correlation_coef)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

# Separate the data points into two groups: local products and cross-border products
local_df = local_df[local_df['price'] < 400000]
local_df = local_df[local_df['total_sales'] < 100000]
cb_df = cb_df[cb_df['price'] < 400000]

local_prices = local_df['price']
local_sales = local_df['total_sales']

crossborder_prices = cb_df['price']
crossborder_sales = cb_df['total_sales']

# Create a scatter plot
plt.scatter(local_prices, local_sales, color='blue', edgecolors='white', label='Local Products')
plt.scatter(crossborder_prices, crossborder_sales, color='orange', edgecolors='white',  label='Cross-border Products')

# Set the labels and title
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Correlation between Price and Total Sales')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
