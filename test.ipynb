```python
# Import packages
import pandas as pd
import os
import numpy as np

# Load data
df = pd.read_csv(os.path.join('../data', 'test.csv'))
df
```

```python
df['Product Category'] = df['Product Category'].fillna(method='ffill')
df
```

```python
transposed_df = pd.pivot_table(df, index='Product Title', columns='Unnamed: 4',
                               aggfunc=lambda x: 1 if x.notna().any() else 0,
                               fill_value=0).reset_index()
# Flatten the column names
transposed_df.columns = ['Product Title'] + [f'Unnamed: 4_{col}' for col in transposed_df.columns[1:]]
transposed_df
```
