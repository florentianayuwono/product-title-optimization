```python
# Import packages
import pandas as pd
import os
import numpy as np
```

```python
# Load data
local_df = pd.read_csv(os.path.join('../data', 'sports_recreation_equipment_local_2000.csv'))
# local_df = local_df.drop(['shop_id', 'item_id', 'source', 'level2_global_be_category', 'total_sales'], axis=1)

cb_df = pd.read_csv(os.path.join('../data', 'sports_recreation_equipment_CB_2000.csv'))
# cb_df = cb_df.drop(['shop_id', 'item_id', 'source', 'level2_global_be_category', 'total_sales'], axis=1)
```

```python
cb_df
```

```python
quality_df = pd.concat([local_df, cb_df], ignore_index=True)
quality_df
```

```python
business_df = pd.read_csv(os.path.join('../data', 'sports_recreation_equipment_business_metrics.csv'))
business_df
```

```python
quality_business_df = pd.merge(quality_df, business_df, on='item_id', how='inner')
quality_business_df
```

```python
rating_df = pd.read_csv(os.path.join('../data', 'sports_recreation_equipment_rating.csv'))
rating_quality_business_df = pd.merge(quality_business_df, rating_df, on='product_title', how='inner')
rating_quality_business_df
```

```python
rating_quality_business_df['ctr_(click/pv)'] = rating_quality_business_df['click'] / rating_quality_business_df['pv']
rating_quality_business_df['uo/pdp_view'] = rating_quality_business_df['uo'] / rating_quality_business_df['pdp_view']
rating_quality_business_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the correlation using seaborn
sns.scatterplot(data=rating_quality_business_df, x='Relevance', y='ctr_(click/pv)')

# Set labels and title
plt.xlabel('Relevancy Rating')
plt.ylabel('CTR')
plt.title('Correlation between Relevancy Rating and CTR')

# Show the plot
plt.show()

correlation_coef = rating_quality_business_df['Fluency'].corr(rating_quality_business_df['ctr_(click/pv)'])
print("Pearson correlation coeff:", correlation_coef)
```
